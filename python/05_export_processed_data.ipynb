# Export Processed Data

This notebook exports the cleaned and enriched dataset for use in Power BI and future analysis.

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/retail_ecommerce_talent_pool_clean.csv")
```

```python
df["total_skills"] = (
    df["excel_skill"] +
    df["sql_skill"] +
    df["python_skill"] +
    df["power_bi_skill"] +
    df["looker_studio_skill"] +
    df["database_skill"] +
    df["tableau_skill"]
)
```

```python
df["skill_level"] = pd.cut(
    df["total_skills"],
    bins=[-1, 2, 5, 7],
    labels=["Beginner", "Intermediate", "Advanced"]
)
```

```python
df[["candidate_id", "total_skills", "skill_level"]].head()
```

```python
df.to_csv(
    "../data/retail_ecommerce_talent_pool_processed.csv",
    index=False
)
```
